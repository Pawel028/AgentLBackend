```python
api_key = os.getenv("google_api_key")
cse_id = os.getenv("cse_id")
print(f"Google API Key: {api_key}")
```

```python
import requests
query = "what was the case of brutal rape and murder in New delhi in 2012?"
def google_search(query, api_key, cse_id, num_results=5):
    url = f"https://www.googleapis.com/customsearch/v1"
    params = {
        'key': api_key,
        'cx': cse_id,
        'q': query,
        'num': num_results
    }
    
    response = requests.get(url, params=params)
    return response.json()
response = google_search(query=query,api_key=api_key, cse_id=cse_id,num_results=2)
```

```python
from bs4 import BeautifulSoup
def get_response_link(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.content, 'html.parser')
    return soup.renderContents()

link_res = get_response_link(response['items'][0]['link'])
```
